##########################################
#### DC RECEIPT TRANSACTION GENERATOR ####
##########################################

```python

# Connect to PostgreSQL
conn = psycopg2.connect(
    dbname="postgres",
    user="postgres",
    password="P0stGr3sM6ntr6",
    host="localhost",
    port="5432"
)
cursor = conn.cursor()

# Parameters for receipts creation
num_receipts = 10  # Number of receipts to create
max_units_received = 100  # Maximum number of units in each receipt

# Get product, location IDs, and stock_on_hand from Item_Loc_SOH for locations with type 'W'
cursor.execute("""
    SELECT product_id, location_id, stock_on_hand
    FROM item_loc_soh
    WHERE location_id IN (SELECT location_id FROM location WHERE location_type = 'W');
""")
records = cursor.fetchall()

# Current date and timestamp for the Event table
cursor.execute("SELECT CURRENT_DATE;")
current_date = cursor.fetchone()[0]

cursor.execute("SELECT CURRENT_TIMESTAMP;")
current_timestamp = cursor.fetchone()[0]

# Create random receipt entries
for _ in range(num_receipts):
    # Select a random product-location record
    product_id, location_id, stock_on_hand = random.choice(records)
    units_received = random.randint(1, max_units_received)

    # Insert a record into the Receipts table
    cursor.execute(
        """
        INSERT INTO Receipts (product_id, location_id, number_of_units_received, receipt_cost)
        VALUES (%s, %s, %s, %s)
        RETURNING receipt_cost
        """,
        (product_id, location_id, units_received, random.uniform(10, 100))
    )
    receipt_cost = cursor.fetchone()[0]

    # Insert a record into the Event table, using stock_on_hand as actual_soh
    cursor.execute(
        """
        INSERT INTO Event (event_type, product_id, location_id, soh_impact, actual_soh, transaction_date, transaction_time)
        VALUES (%s, %s, %s, %s, %s, %s, %s)
        """,
        ('receipt', product_id, location_id, units_received, stock_on_hand, current_date, current_timestamp)
    )

    # Update stock_on_hand in Item_Loc_SOH table
    cursor.execute(
        """
        UPDATE Item_Loc_SOH
        SET stock_on_hand = stock_on_hand + %s
        WHERE product_id = %s AND location_id = %s
        """,
        (units_received, product_id, location_id)
    )

# Commit and close
conn.commit()
cursor.close()
conn.close()
```

##########################################
####  ALLOCTION TRANSACTION GENERATOR ####
##########################################

```python

# Connect to PostgreSQL
conn = psycopg2.connect(
    dbname="postgres",
    user="postgres",
    password="P0stGr3sM6ntr6",
    host="localhost",
    port="5432"
)
cursor = conn.cursor()

# Step 1: Select product_id, location_id, and stock_on_hand for warehouses (W) where stock_on_hand > 0
cursor.execute("""
    SELECT product_id, location_id, stock_on_hand
    FROM item_loc_soh
    WHERE location_id IN (
        SELECT location_id FROM location WHERE location_type = 'W'
    ) AND stock_on_hand > 0
    LIMIT 20;
""")
warehouse_records = cursor.fetchall()

# Get store location IDs (location_type = 'S')
cursor.execute("""
    SELECT location_id FROM location WHERE location_type = 'S';
""")
store_locations = [loc[0] for loc in cursor.fetchall()]

# Current date and timestamp for the Event table
cursor.execute("SELECT CURRENT_DATE;")
current_date = cursor.fetchone()[0]

cursor.execute("SELECT CURRENT_TIMESTAMP;")
current_timestamp = cursor.fetchone()[0]

# Step 2: Move inventory from a warehouse to a store
for product_id, warehouse_location_id, stock_on_hand in warehouse_records:
    # Determine a random quantity to move (less than or equal to stock_on_hand)
    units_to_move = random.randint(1, stock_on_hand)

    # Select a random store location to move inventory to
    store_location_id = random.choice(store_locations)

    # Decrease stock_on_hand in the warehouse (W)
    cursor.execute(
        """
        UPDATE item_loc_soh
        SET stock_on_hand = stock_on_hand - %s
        WHERE product_id = %s AND location_id = %s
        """,
        (units_to_move, product_id, warehouse_location_id)
    )

    # Increase stock_on_hand in the store (S), or create a record if it doesn't exist
    cursor.execute(
        """
        INSERT INTO item_loc_soh (product_id, location_id, stock_on_hand, average_cost, reserved_quantity, non_sellable_quantity)
        VALUES (%s, %s, %s, (SELECT average_cost FROM item_loc_soh WHERE product_id = %s AND location_id = %s), 0, 0)
        ON CONFLICT (product_id, location_id)
        DO UPDATE SET stock_on_hand = item_loc_soh.stock_on_hand + EXCLUDED.stock_on_hand
        """,
        (product_id, store_location_id, units_to_move, product_id, warehouse_location_id)
    )

    # Step 3: Insert events for the allocation
    # Event for the warehouse (allocation_shipment)
    cursor.execute(
        """
        INSERT INTO Event (event_type, product_id, location_id, soh_impact, actual_soh, transaction_date, transaction_time)
        VALUES (%s, %s, %s, %s, %s, %s, %s)
        """,
        ('allocation_shipment', product_id, warehouse_location_id, -units_to_move, stock_on_hand - units_to_move, current_date, current_timestamp )
    )

    # Event for the store (allocation_receipt)
    cursor.execute(
        """
        INSERT INTO Event (event_type, product_id, location_id, soh_impact, actual_soh)
        VALUES (%s, %s, %s, %s, (SELECT stock_on_hand FROM item_loc_soh WHERE product_id = %s AND location_id = %s))
        """,
        ('allocation_receipt', product_id, store_location_id, units_to_move, product_id, store_location_id)
    )

# Commit and close
conn.commit()
cursor.close()
conn.close()
```
