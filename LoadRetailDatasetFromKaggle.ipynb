```python
import zipfile
import pandas as pd

with zipfile.ZipFile(download_path + 'retail-products-classification.zip', 'r') as zip_ref:
    zip_ref.extractall(download_path)
```

```python


# Load the dataset
df_initial = pd.read_csv(download_path + 'train.csv')
df_initial = df_initial.drop(columns=['ImgId'])
df_initial.rename(columns = {'categories' : 'category_name'}, inplace = True)

# Display the first few rows
df_initial.head()
```

```python
# Extract unique categories
df_categories = pd.DataFrame(df_initial['categories'].unique(), columns=['category_name'])

# Assign a category_id starting from 1000
df_categories['category_id'] = range(1000, 1000 + len(df_categories))
df_categories['comment'] = ''
# Display the unique categories dataframe
df_categories.head()
```

```python
df_categories.head()
```

```python
# Select title and description columns, and drop duplicates to get unique products
df_product = df_initial[['title', 'description', 'category_name']].drop_duplicates().reset_index(drop=True)

# Assign product_id starting from 30000000
df_product['product_id'] = range(30000000, 30000000 + len(df_product))

# Merge with df_categories to get category_id
df_product = df_product.merge(df_categories, on='category_name', how='left')

# Drop the original 'category' column as category_id is now added
df_product = df_product.drop(columns=['category_name'])

# Display the resulting df_product DataFrame
df_product.head()
```

```python
import random
import numpy as np
# Categories and their associated ranges
category_ranges = {
    "range_10_100": [1000, 1003, 1004, 1006, 1007, 1020],
    "range_1000_20000": [1001, 1010, 1011, 1016, 1012, 1017, 1018],
    "range_100_5000": [1002, 1008, 1014, 1007, 1009, 1020, 1019]
}

# Set average_cost based on category ranges
def set_average_cost(row):
    if row['category_id'] in category_ranges["range_10_100"]:
        return round(random.uniform(10, 100), 2)
    elif row['category_id'] in category_ranges["range_1000_20000"]:
        return round(random.uniform(1000, 20000), 2)
    elif row['category_id'] in category_ranges["range_100_5000"]:
        return round(random.uniform(100, 5000), 2)
    else:
        return round(random.uniform(100, 2000), 2)

# Calculate the unit_retail with a 70% markup
def calculate_unit_retail(average_cost):
    markup_price = average_cost * 1.70  # Apply a 70% markup
    return round(markup_price * 4) / 4  # Round to nearest 0.05 (common tender value)

# Apply the functions to set average_cost and calculate unit_retail
df_product['average_cost'] = df_product.apply(set_average_cost, axis=1)
df_product['unit_retail'] = df_product['average_cost'].apply(calculate_unit_retail)
df_product['promotional_price'] = np.NaN
```

```python
from faker import Faker

# Initialize Faker with Indian locale
fake = Faker('en_IN')

# Generate 4 warehouses with location_type 'W'
warehouses = []
for i in range(1, 5):
    warehouses.append({
        'location_id': f"{i:02}",  # 2-digit location number
        'location_name': f"Warehouse {i}",
        'location_type': 'W',
        'address': fake.address()
    })

# Generate 20 stores with location_type 'S'
stores = []
for i in range(5, 25):
    stores.append({
        'location_id': f"{(i+2000):04}",  # 4-digit location number
        'location_name': f"Store {i - 5}",
        'location_type': 'S',
        'address': fake.address()
    })

# Combine warehouses and stores into one list
locations = warehouses + stores

# Convert to a DataFrame
df_locations = pd.DataFrame(locations)

# Display the DataFrame
df_locations
```

```python
for _, row in df_locations.iterrows():
    cursor.execute(
        """
        INSERT INTO Location (location_id, location_name, location_type, address)
        VALUES (%s, %s, %s, %s)
        """,
        (row['location_id'], row['location_name'], row['location_type'], row['address'])
    )

# Commit the transaction and close the connection
conn.commit()
cursor.close()
conn.close()
```
